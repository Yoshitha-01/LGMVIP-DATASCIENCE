```python
from google.colab.patches import cv2_imshow
from skimage import io
import cv2
```

```python
img=io.imread("https://laverdadnoticias.com/img/2020/07/25/bts_kim_tae_hyung_v.jpg")
cv2_imshow(img)
```

```python
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img2)
```

```python
img3=cv2.bitwise_not(img2)
cv2_imshow(img3)
```

```python
img4=cv2.GaussianBlur(img2,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(img4)
```

```python
final=cv2.divide(img2,255-img4,scale=250)
cv2_imshow(final)
```
